```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tvs = pd.read_csv('D:/upgrad/ML/tvmarketing.csv')
```

```python
tvs = tvs[['TV','Sales']]
```

```python
tvs.head()
```

```python
tvs.describe()
```

```python
plt.style.use('ggplot')
sns.regplot(tvs.TV,tvs.Sales)
```

```python
X = tvs['TV']
```

```python
X.head()
```

```python
Y=tvs['Sales']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size=0.7, test_size=0.3, random_state=100)
```

```python
print(len(X_train))
print(len(Y_train))
print(len(X_test))
print(len(Y_test))
```

```python
# Since we have a Series we need to convert it to 2-D array to use with scikit learn

X_train = X_train[:,np.newaxis]
X_test = X_test[:,np.newaxis]

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
Y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(Y_test, Y_pred)
r2s = r2_score(Y_test, Y_pred)
```

```python
print('Mean Squared Error', mse)
print('R squared', r2s)
```

```python
c = [i for i in range(1,61,1)]
```

```python
fig = plt.figure()
plt.plot(c, Y_test, color="blue", linewidth="2.5", linestyle="-")
plt.plot(c, Y_pred, color="red", linewidth="2.5", linestyle="-")
fig.suptitle('Actual and Predicted', fontsize=20)
plt.xlabel('Index',fontsize=16)
plt.ylabel('Sales',fontsize=18)
```

```python
fig = plt.figure()
plt.plot(c, Y_test-Y_pred, color="blue", linewidth="2.5", linestyle="-")
fig.suptitle('Error Terms', fontsize=20)
plt.xlabel('Index',fontsize=16)
plt.ylabel('Ytest - Ypred',fontsize=18)
```

```python
# Check distribution of error

sns.distplot(Y_test-Y_pred)
plt.xlabel('Index', fontsize=18)
plt.ylabel('Error', fontsize=16)
```
